```python
#Selection by label
#loc
#pandas.DataFrame.loc
#Access a group of rows and columns by label(s) or a boolean array
#loc[] is primarily label based, but may also be used with a boolean array
#Allowed inputs are:
#   1-A single label, e.g. 5 or 'a', (note that 5 is interpreted as a label of the index,
#and never as an integer position along the index).
#   2-A list or array of labels, e.g. ['a', 'b', 'c'].
#   3-A slice object with labels, e.g. 'a':'f'.
#   4-A boolean array of the same length as the axis being sliced, e.g. [True, False, True].
#   5-A callable function with one argument (the calling Series or DataFrame)
#and that returns valid output for indexing (one of the above)

import pandas as Behdad
ColumData1={"Measure1":["a1", [1,2], "Creating Data", "Jam", 24, 78.9, 31, 600],\
           "Measure2": ["ZZ","WW", [23, 1], "BB24", 43, 1.12, 786, 3],\
           "Measure3":[12, 3, 2, 8.999999, 4.8, 9, 45, 76], "Measure4":[1232, 3432, 788,46554, 324, 456, 5, 2],\
           "Measure5": ["er","WW", 34, "RRFF6", "WW", 94,"WW", 6],"Measure6": ["34ZZ","43WW", "True4", "24SS", 645, 4, 32, "43WW"],\
           "Measure7":[12, 454, 7, 95, 32, 1, 9, 8]}

#Single label. Note this returns the row as a Series
ColumData12=Behdad.DataFrame(ColumData1)
LOC3=ColumData12.loc[1]
print(LOC3)

ColumData12=Behdad.DataFrame(ColumData1)
LOC3=ColumData12.loc[2]
print(LOC3)

#List of labels. Note using [[]] returns a DataFrame
print("\n using [[]] returns a DataFrame")

ColumData12=Behdad.DataFrame(ColumData1)
LOC3=ColumData12.loc[[0, 2, 3]]
print(LOC3)

ColumData12=Behdad.DataFrame(ColumData1)
LOC3=ColumData12.loc[[0, 2, 3]]
print(LOC3)

#Single label for row and column
print("\nSingle label for row and column")
ColumData12=Behdad.DataFrame(ColumData1)
LOC3=ColumData12.loc[1, "Measure2"]
print(LOC3)

print("\nSingle label for row and column")
ColumData12=Behdad.DataFrame(ColumData1)
LOC3=ColumData12.loc[1, "Measure3"]
print(LOC3)

print("\nSingle label for row and column")
ColumData12=Behdad.DataFrame(ColumData1)
LOC3=ColumData12.loc[3, "Measure5"]
print(LOC3)

#Slice with labels for row and single label for column
#As mentioned above, note that both the start and stop of the slice are included
print("\n Slice with labels Ex1:","\n 1:3")
ColumData12=Behdad.DataFrame(ColumData1)
LOC3=ColumData12.loc[1:3, "Measure2"]
print(LOC3)

print("\n Slice with labels Ex2:", "\n 0:3")
ColumData12=Behdad.DataFrame(ColumData1)
LOC3=ColumData12.loc[0:3, "Measure3"]
print(LOC3)

#Boolean list with the same length as the row axis
print('\n Boolean list with the same length as the row axis Ex1:')
ColumData12=Behdad.DataFrame(ColumData1)
LOC3=ColumData12.loc[[True, False, True, False, True, False, True, False]]
print(LOC3)
print('\n Boolean list with the same length as the row axis Ex2:')
ColumData12=Behdad.DataFrame(ColumData1)
LOC3=ColumData12.loc[[True, False, False, False, True, False, True, False]]
print(LOC3)

#Conditional that returns a boolean Series
print('\n Conditional that returns a boolean Series Ex 1:', '\n Measure4>1000')
cond=ColumData12.loc[ColumData12['Measure4'] > 1000]
print(cond)

print('\n Conditional that returns a boolean Series Ex 2:', '\n Measure4<1000')
cond=ColumData12.loc[ColumData12['Measure4'] < 1000]
print(cond)

print('\n Conditional that returns a boolean Series Ex 2:', '\n Measure6==43WW')
cond=ColumData12.loc[ColumData12['Measure6']=="43WW"]
print(cond)

#Conditional that returns a boolean Series with column labels specified
print('\n Conditional that returns a boolean Series with column labels specified Ex1')
cond=ColumData12.loc[ColumData12['Measure4'] ==788, ['Measure6']]
print("\n Measure4==788 \n",cond)

print('\n Conditional that returns a boolean Series with column labels specified Ex1')
cond=ColumData12.loc[ColumData12['Measure3'] >=9, ['Measure6', 'Measure3', 'Measure1', 'Measure4']]
print("\n Measure3 >=9 \n",cond)

print('\n Conditional that returns a boolean Series with column labels specified Ex3')
cond=ColumData12.loc[ColumData12['Measure1'] =="Creating Data", ['Measure2', 'Measure1']]
print("\n 'Measure1==Creating Data \n", cond)

print('\n Conditional that returns a boolean Series with column labels specified Ex4')
cond=ColumData12.loc[ColumData12['Measure5'] !="WW", ['Measure2', 'Measure4', 'Measure5']]
print('\nMeasure5 !=WW',cond)


#Callable that returns a boolean Series
print('\n Callable that returns a boolean Series Ex1')
ColumData12.loc[lambda ColumData12: ColumData12['Measure2'] == 43]
print('\nMeasure2 ==43',cond)     

```
